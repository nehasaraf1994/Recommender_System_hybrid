```python
import pandas as pd
import gzip
from PIL import Image
from PIL import ImageOps
import urllib.request as urllib
import urllib.error as err

import io
import numpy as np

def parse(path):
    g = gzip.open(path, 'rb')
    for line in g:
        yield eval(line)

def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')

df = getDF('reviews_Clothing_Shoes_and_Jewelry_5.json.gz')
```

#### Load encoder2.h5 model

```python
import numpy as np
from keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D
from keras.models import Model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
from keras.models import load_model
encoder = load_model('encoder2.h5')
```

#### Reads each image url, creates the image array and returns its vector representation using encoder model

```python
def get_image_array(url):
    if isinstance(url, str):
        try:
            fd = urllib.urlopen(url)
            image_file = io.BytesIO(fd.read())
            im = Image.open(image_file)
            size = 28, 28
            img = im.resize(size, Image.ANTIALIAS)
            img = img.convert(mode="RGB")
            print(np.array(img).shape)
            image = encoder.predict(np.array(img).reshape(1,28,28,3))
            image = image.reshape(1,4*4*1)
            vector = image[0].tolist()
            return ",".join(map(str, vector))
        except err.HTTPError as error:
            return "None"
    
```

#### Create dataframe of all unique items in reviews dataset

```python
items= df['asin'].unique().tolist()
```

```python
df2_items = pd.DataFrame(columns=["asin"])
df2_items["asin"] = items
df2_items.head(5)
```

#### Apply get_image_array function to each item and add a column of image vectors

```python
df2_items['img_vector'] = df2_items.apply (lambda row: get_image_array(df2.loc[df2['asin'] == row["asin"], 'imUrl'].iloc[0]),axis=1)
```

#### Store the vector representation of each image in review dataset

```python
df2_items.to_csv("item_vector_image3.csv")
```
